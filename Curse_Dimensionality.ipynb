# Curse of Dimensionality

  En Machine Learning, el número de dimensiones se puede equiparar al número de variables o características (features) que estemos utilizando.

La maldición de la dimensión se «manifiesta» de dos maneras:

1. **La distancia media entre los datos aumenta con el número de dimensiones:** A medida que el número de dimensiones aumenta, la distancia media entre ellos también aumenta. Esto en sí mismo no es muy problemático para el aprendizaje automático. 

2.**La variabilidad de la distancia disminuye exponencialmente con el número de dimensiones:** Esto es terrible para los métodos de aprendizaje automático basados en distancias tales como clustering o los vecinos más cercanos. El problema es que cuando hay un gran número de atributos (features), los datos están todos a casi la misma distancia. Es decir, no hay variabilidad entre sus distancias.


```
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
```

# One Dimension

```
d1 = np.random.rand(10,5)
d1
```

```
poly1 = PolynomialFeatures(3)
pd.DataFrame(poly1.fit_transform(d1))
```

Second Dimension

```
d2 = np.random.rand(10,6)
d2
```

```
poly2 = PolynomialFeatures(3)
pd.DataFrame(poly1.fit_transform(d2))
```

Third Dimension

```
d3 = np.random.rand(10,7)
d3
```

```
poly3 = PolynomialFeatures(3)
pd.DataFrame(poly1.fit_transform(d3))
```

# Conclusion: 

 Como podemos ver en la grafica de abajo a medida que el número de dimensiones crece los puntos se alejan unos de otros, y como resultado el tamaño de cada neighborhod (el espacio de los puntos más cercanos a un punto de interés) aumenta.

```
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.title(
    "% volume covered by $r$ uniform samples on a $p$-hypercube"
)
plt.xlabel('$p$')
plt.ylabel('% volume covered')

plt.plot(range(1, 10), [0.1**(1/n) for n in range(1, 10)], label="r = 10%")
plt.plot(range(1, 10), [0.2**(1/n) for n in range(1, 10)], label="r = 20%")
plt.plot(range(1, 10), [0.33**(1/n) for n in range(1, 10)], label="r = 33%")
plt.plot(range(1, 10), [0.5**(1/n) for n in range(1, 10)], label="r = 50%")
plt.plot(range(1, 10), [0.8**(1/n) for n in range(1, 10)], label="r = 80%")
plt.legend()
```

Entre más features poseamos antes de generar un polinomio, mas features tendremos luego de generar un nuevo polinomio

```
print('Combinando 1 Dimension:',np.shape(pd.DataFrame(poly1.fit_transform(d1)))[1])
print('Combinando 2 Dimension:',np.shape(pd.DataFrame(poly2.fit_transform(d2)))[1])
print('Combinando 3 Dimension:',np.shape(pd.DataFrame(poly3.fit_transform(d3)))[1])
```
